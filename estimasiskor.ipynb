```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# memanggil dataset

```python
df = pd.read_csv('understat.com.csv')
```

# visualisasi data

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(),annot=True)
```

# jumlah matches berdasarkan team

```python
models = df.groupby('team').count()[['matches']].sort_values(by='matches',ascending=True).reset_index()
models = models.rename(columns={'matches':'number Of team'})
```

```python
fig = plt.figure(figsize=(15,5))
sns.barplot(x=models['team'], y=models['number Of team'], color='royalblue')
plt.xticks(rotation=60)
```

# distribusi jumlah skor

```python
plt.figure(figsize=(15,5))
sns.distplot(df['scored'])
```

# seleksi fitur

```python
features = ['matches','wins','draws','loses']
x = df[features]
y = df['scored']
x.shape, y.shape
```

# split data training dan data testing

```python
from sklearn.model_selection import train_test_split
x_train, X_test, y_train, y_test = train_test_split(x,y,random_state=70)
y_test.shape
```

# membuat model regresi linear

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
pred = lr.predict(X_test)
```

```python
score = lr.score(X_test, y_test)
print('akurasi model regresi linier = ', score)
```

# membuat inputan model regresi linear

```python
#matches = 38, wins = 22, draws = 11, loses = 5                                                                                                                                                                                                                                                                                              
input_data = np.array([[38,22,11,5]])

prediction = lr.predict(input_data)
print('Estimasi Jumlah Skor :', prediction)
```

```python
import pickle

filename = 'skor.sav'
pickle.dump(lr,open(filename,'wb'))
```
